```python
# Linear Regression: To fit the model to the given data. 
#  Import the data into a dataset array. 
#  Create a fitted model. 
#  Locate and remove outliers. 
#  Simplify the model. 
#  Predict responses to new data. 
#  Share the model.
# The population (p) of a small community on the outskirts of a city grows rapidly over a 20period: 
# t 0 5 10 15 20 
# p 100 200 450 950 2000 
# As an engineer working for a utility company, you must forecast the population 5 years intofuture in order to anticipate the demand for power. 
# Employ a linear regression model to makeprediction. Predict the population of the community after 25 years

import matplotlib.pyplot as plt

t=[0,5,10,15,20] 
p=[100,200,450,950,2000]

def mean(c):
    sum=0
    for i in range(len(c)):
        sum=sum+c[i]
    return (sum/len(c))
mean_t=mean(t)
print("Mean of t:",mean_t)
mean_p=mean(p)
print("Mean of p:",mean_p)

#x=x_minus_xbar_sq
def sum_x_min_xbarsq(c):
    x=[]
    for i in range(0,len(c)):
        x.append((c[i]-mean(c))*(c[i]-mean(c)))
    #print(x)
    sum_x=0
    for i in range(0,len(x)):
        sum_x=sum_x+x[i]
    return sum_x

sum_t_min_tbarsq=sum_x_min_xbarsq(t)
print("Sigma ti Minus tbar whole Square is : ",sum_t_min_tbarsq)
sum_p_min_pbarsq=sum_x_min_xbarsq(p)
print("Sigma pi Minus pbar whole Square is : ",sum_p_min_pbarsq)

#tp=t_min_xbar_into_p_min_pbar
def sumof_t_min_tbar_into_p_min_pbar():
    tp=[]
    for i in range(0,len(t)):
        tp.append((t[i]-mean_t)*(p[i]-mean_p))
    #print(tp)
    sum_tp=0
    for i in range(0,len(tp)):
        sum_tp=sum_tp+tp[i]
    return sum_tp

sumof_t1p1=sumof_t_min_tbar_into_p_min_pbar()
print("Sigma ti Minus tbar into pi Minus pbar is : ",sumof_t1p1)

#Regression Coefficient are
btp=sumof_t1p1/sum_p_min_pbarsq
bpt=sumof_t1p1/sum_t_min_tbarsq
print("Regression Coefficient are:",bpt,"and",btp)

print("Regression line of y on x: y=",mean_p+(bpt*mean_t),"+",bpt,"x")
x1=[0,5,10,15,20]
y1=[]
for i in range(len(x1)):
    y1.append(mean_p+(bpt*(x1[i]-mean_t)))

print("Regression line of x on y: x=",mean_t+(btp*mean_p),"+",btp,"y")
x2=[]
y2=[100,200,450,950,2000]
for i in range(len(y2)):
    x2.append(mean_t+(btp*(y2[i]-mean_p)))

#Predict the population of the community after 25 years
print("The population after year 25 is ",mean_p+bpt*(25 - mean_t) )
#print("The population after year 25 is ",(25-mean_t+(btp*mean_p))/btp)


plt.plot(x1,y1,label='Regression line of p on t',color='red')
plt.plot(x2,y2,label='Regression line of t on p')
plt.title('Multiple Regression Lines')
plt.xlabel('t')
plt.ylabel('p')
plt.scatter(t,p,label='Original Data')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[14,27,40,55,68]
n=len(x)
def sum_of_list(c):
    sum = 0
    for i in range(len(c)):
        sum += c[i]
    return sum

sigma_x=sum_of_list(x)
sigma_y=sum_of_list(y)

def sum_of_xsq(c):
    sq_sum=0
    for i in c:
        sq_sum=sq_sum+(i*i)
    return sq_sum
    
sum_of_xsq=sum_of_xsq(x)
#print(sum_of_xsq)

def sum_of_xy():
    xy_sum=0
    for i in range(n):
        xy_sum=xy_sum+(x[i]*y[i])
    return xy_sum
sum_of_xy=sum_of_xy()
#print(sum_of_xy)

def solve_linear_system(A, B):
    n = len(A)
    for i in range(n):
        for j in range(i + 1, n):
            ratio = A[j][i] / A[i][i]
            for k in range(i, n):
                A[j][k] -= ratio * A[i][k]
            B[j] -= ratio * B[i]
    X = [0] * n
    for i in range(n - 1, -1, -1):
        X[i] = B[i] / A[i][i]
        for j in range(i - 1, -1, -1):
            B[j] -= A[j][i] * X[i]
    return X

#normal equations
print("First Equation:",n,"a +",sigma_x,"b =",sigma_y)
print("Second Equation:",sigma_x,"a +",sum_of_xsq,"b =",sum_of_xy)

A = [[n,sigma_x],[sigma_x,sum_of_xsq]]
B = [sigma_y,sum_of_xy]
solution = solve_linear_system(A, B)
print("Final Equation: y=",solution[0],"+",solution[1],"x")

xpoint = x
ypoint=[]
for i in range(len(xpoint)):
    y1=solution[0]+(solution[1]*xpoint[i])
    ypoint.append(y1)


plt.scatter(x, y,label='Original Data')
plt.plot(xpoint,ypoint,label='Best Fit')
plt.title('Fitting a Straight Line')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
#A large circular dartboard is set up with a bulls eye at the center of the circle, which is at co
#ordinate (0,0). A dart is thrown at the center but lands at (X, Y) where X and Y are two
#different Gaussian Random variables. Write a program to find the average distance of the d
#art from the bulls eye for 5, 10, 1000, 1000000 iterations. Plot the scatter diagram for the same.
import random
import matplotlib.pyplot as plt

i=int(input("Enter number of iteration:"))
x=[]
y=[]
for j in range(0,i):
    x.append(round(random.uniform(-5,5),2))
    y.append(round(random.uniform(-5,5),2))
#print(x)
#print(y)

#distance
distance=[]
for j in range(i):
    distance.append(((x[j]*x[j])+(y[j]*y[j]))**0.5)
#print(distance)

#Average
avg=0
for j in range(i):
    avg=(distance[j]+avg)/i
print("The average distance of the dart from the bulls eye is:",avg)
    
# Plot the scatter diagram
plt.figure(figsize=(500, 500))
plt.title('Dart Throwing Simulation')
plt.scatter(0,0,color='red')
plt.scatter(x,y,label='scatter')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulate for different iteration counts
iterations = [5, 10, 1000, 1000000]

# Function to calculate average distance
def average_distance(num_iterations):
    # Generate random (X,Y) coordinates 
    X = np.random.normal(0, 1, num_iterations)
    Y = np.random.normal(0, 1, num_iterations)
    
    # Calculate the Euclidean distance from the bullseye (0, 0)
    distances = np.sqrt(X**2 + Y**2)
    avg_distance = np.mean(distances)

    # Plot the scatter diagram for the dartboard
    plt.figure(figsize=(4, 4))  # Create a square figure
    
    # Add concentric colored circles
    circle_radii = [5, 4, 3, 2, 1]  # radius for the rings(biggest to smallest)
    circle_colors = ['white', 'black', 'blue', 'red','yellow']  #colors of rings
    
    for radius, color in zip(circle_radii, circle_colors):
        circle = plt.Circle((0, 0), radius, color=color, fill=True, ec='black', lw=2)
        plt.gca().add_patch(circle)  # Add the circle to the current axes
    
    # Scatter the dart positions
    plt.scatter(X, Y, edgecolor='k')
    plt.scatter(0,0,edgecolor='black',color='red')
    
    #plot
    plt.axhline(0, color='black',linewidth=1)  # horizontal line through origin
    plt.axvline(0, color='black',linewidth=1)  # vertical line through origin
    plt.title("Scatter Diagram for Darts")
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.grid(True)
    plt.gca().set_aspect('equal', adjustable='box')  # Keep the aspect ratio equal
    plt.xlim(-6, 6)  # Limit axes for better visualization
    plt.ylim(-6, 6)
    
    plt.show()
    return avg_distance

# Calculate and display average distances
for n in iterations:
    avg_dist = average_distance(n)
    print('Average distance for', n ,' iterations: ',avg_dist)
```
