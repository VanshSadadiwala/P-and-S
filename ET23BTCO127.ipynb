```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('https://media.geeksforgeeks.org/wp-content/uploads/tips.csv')
plt.scatter(df['total_bill'],df['tip'])
plt.title('Scatter Plot')
plt.xlabel('total_bill')
plt.ylabel('tip')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('https://media.geeksforgeeks.org/wp-content/uploads/tips.csv')
plt.plot(df['sex'],df['total_bill'])
plt.xlabel("sex")
plt.ylabel("total bill")
plt.title("Line chart")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('https://media.geeksforgeeks.org/wp-content/uploads/tips.csv')
plt.bar(df['day'],df['total_bill'],color='yellow')
plt.xlabel("day")
plt.ylabel("total bill")
plt.title("Line chart")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('https://media.geeksforgeeks.org/wp-content/uploads/tips.csv')
plt.hist(df['total_bill'],color='orange')
plt.xlabel("x axis")
plt.ylabel("total bill")
plt.title("Line chart")
plt.show()
```
